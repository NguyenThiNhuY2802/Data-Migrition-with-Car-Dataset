```python
import pandas as pd

import sys
```

## 1. Load data from CSV to dataframe

```python
df = pd.read_csv("car_sample_data.csv")
df.info()
df = pd.read_csv("car_sample_data.csv", na_values="n/a")
df.info()
df.head()
```

## B. Transfer
## 1. Check NA  "price" column

```python

df_NA = df.dropna(subset = ['Price']) # Xác định cụ thể cột cần kiểm tra và loại bỏ nếu có chứa ô trống
df_NA
```

## 2. Check length  "drive-wheels": if >5 then remove

```python
df_index = df_NA.loc[df_NA['Drive-wheels'].str.len()>5].index
print(df_index) # tìm drive-wheels có độ dài lớn hơn 5
df_NA.drop(df_index)
df_length = df_NA
df_length
```

## 3. Check "Height, Width, Length, Engine-size, Bore, Stroke, Price" datatype: if not interger then remove

```python
List_col = ['Height','Width', 'Length', 'Engine-size', 'Bore', 'Stroke', 'Price']
L = []
for i in List_col:
    if df_length[i].dtypes != 'int64':
        L.append(i)
df_datatype = df_length.drop(List_col, axis = 1)
df_datatype

```

## 4. Check duplicate

```python
df_dup = df_datatype.drop_duplicates()
df_dup
```

## 5. Rename column "Fuel-type" to "type", "Num-of-cylinders" to  "Num_cylinders"

```python
df_rename = df_datatype.rename(columns ={'Fuel-type':'type', 'Num-of-cylinders':'Nym_cylinders'})
df_rename
```

## 6. Anonimize "dayUpdate" column to "1"

```python
df_rename ['Load-date'] = df_rename ['Load-date'].astype('datetime64')
df_anonimize = df_rename ['Load-date'].apply(lambda x: x.replace(day=1))
df_anonimize 
```
